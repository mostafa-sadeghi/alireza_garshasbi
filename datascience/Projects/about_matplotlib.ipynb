```python
cars_sales["Price"]  = cars_sales["Price"].str.replace('[\$\,\.]','').astype(int)
```

```python
cars_sales
```

```python
# cars_sales["Sale date"] = pd.date_range("1/1/2023",periods =10,freq = "3M")
cars_sales["Sale date"] = pd.date_range("1/1/2023", periods= len(cars_sales))
cars_sales
```

```python
cars_sales["total sales"] = cars_sales["Price"].cumsum()
cars_sales
```

```python
cars_sales.plot(x = "Sale date", y= "total sales");
sum_of_all= sum(cars_sales["Price"])
x = []
for _ in cars_sales["Price"]:
    x.append(_/sum_of_all)
    
cars_sales["our new column"] = x
cars_sales
```
