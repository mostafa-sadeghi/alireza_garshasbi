```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import preprocessing

data = pd.read_csv('car.data')
data.head()
```

```python
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
buying
```

```python
maint = le.fit_transform(list(data["maint"]))
doors = le.fit_transform(list(data["doors"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
target = le.fit_transform(list(data["target"]))

# doors = list(data['doors'])
# doors = [int(door) for door in doors] # if just decimal values exists in string 
```

```python

X = list(zip(buying, maint,doors, persons,lug_boot,safety))
```

```python
X
```

```python
y = list(target)
y
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y,test_size=0.1)

X_train
```

```python
model = KNeighborsClassifier(n_neighbors=7)
model.fit(X_train,y_train)
accuracy = model.score(X_test, y_test)
accuracy
```

```python
predicted = model.predict(X_test)
predicted
X_test


```

```python
names = ['unacc', 'acc', 'good', 'vgood']
for x in range(len(predicted)):
    print("Predicted: ",names[predicted[x]], "Data:", X_test[x], "Actual class: ", names[y_test[x]])
    n = model.kneighbors([X_test[x]], 9, True)
    print(n[0])
    
```
